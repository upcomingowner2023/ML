```python
df.drop('Paymen Type', axis=1, inplace=True)
df.shape
```

```python
df.columns
```

```python
# Convert date column to datetime data type
df['Date'] = pd.to_datetime(df['Date'])

# Set date column as the index of the DataFrame
df.set_index('Date', inplace=True)

# Group the data by month and count the number of data points in each month
monthly_counts = df.resample('M').count()

monthly_counts['Total'].head()
```

```python
yearly_counts = df.resample('Y').count()
yearly_counts['Total'].head()
```

```python
# Reset the index and create a new 'Date' column
df = df.reset_index()
```

```python
# Group the data by month and aggregate the total sales for each month
df_monthly_sales = df.groupby(pd.Grouper(key='Date', freq='M'))['Total'].sum()
df_monthly_sales
```

By seeing the data we can conclude that , In 1st month the sales are high

```python
# Group the data by year and aggregate the total sales for year
df_yearly_sales = df.groupby(pd.Grouper(key='Date', freq='Y'))['Total'].sum()
round(df_yearly_sales)
```

Total sales of the year is = **322967** rupees.

**Aggregating Functions:**

```python
# Calculate the total sales for each product line
df_product_sales = df.groupby('Product line')['Total'].sum()
round(df_product_sales)
```

We can conclude that most of the customers sales are done on Food and beverages.

```python
# Calculate the average rating for each product line
df_product_rating = df.groupby('Product line')['Rating'].mean()
df_product_rating
```

Highest rating of the product is **"Food and beverages"**

**Manipulating Functions:**


```python
# Add a new column with the cost of goods sold (COGS) for each record
df['COGS'] = df['Quantity'] * df['Cost Price']
df['COGS']
```

**Filtering (loc, iloc):**

```python
# Select all records where the customer rating is above 4
df_high_rating = df.loc[df['Rating'] > 4]
df_high_rating
```

It means that 989 customers are 50% satisfied with thier products
